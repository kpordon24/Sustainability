##  Steel Dynamics Sustainability: Executive Level Benchmarking & Future Strategies

## Project Recap and Key Findings

First, let us revisit the business problem and then answer each question based on our findings.

Business Problem: The Steel Dynamics board seeks to enhance their environmental, social, and governance (ESG) performance and manage risks. They need insights on: (1) their current ranking within the U.S. steel industry, (2) which KPIs will deliver the most tangible and intangible benefits, and (3) requirements for becoming a leader in the industry.

### 1. What is Steel Dynamics current ranking within the U.S. steel industry?

Since 2010, Steel Dynamics has consistently ranked at the bottom of the industry in both financial performance and ESG rating, as clearly demonstrated during the exploratory data analysis stage (see graphs below). While the company has shown long-term improvement, this progress has been outpaced by competitors, as the entire industry has experienced steady advancement during this period. In the following two questions, we will explain why this is and how the Steel Dynamics board can improve its positioning.

## 2. Which KPIs will deliver the most tangible and intangible benefits?

To address this question, we evaluated each attribute's capacity to influence both tangible outcomes (revenue and market capitalization) and intangible metrics (ESG scores). Our process involved developing three distinct machine learning models (Lasso, Random Forest, and XGBoost), which were then evaluated based on their ability to explain variation and predict outcomes using unseen data (see graphs below).

<b>Revenue:</b> The set of graphs above shows that XGB performed best at explaining and predicting revenue, with R-squared and F1 scores of 83%.

<b>ESG Score:</b> The linear Lasso model was best at explaining the variation in and predicting ESG score with R-squared and F1 scores of 98% and 100%, respectively.

<b>Market Cap:</b> Random Forest was the best at predicting market capitalization. However, most of the variation was left unexplained by the model, suggesting our dataset is missing important metrics such as consumer confidence.

## 3. How can Steel Dynamics becoming an industry leader?

The analysis yields a clear conclusion supported by strong evidence: to enhance its sustainability positioning, the board should prioritize the KPIs. For instance, increasing the use of recycled steel, water, and renewable energy is projected to improve Steel Dynamics' ESG score, a finding backed by the strong performance of all three machine learning models. Conversely, while increased sustainability investments, R&D spending, and renewable energy use might positively impact market capitalization, model performance regarding this outcome was considerably weaker, indicating less predictable results.

Lastly, our work demonstrated a noteable relationship between specific ESG metrics and a company's revenue position. Enhanced sustainability and biodiversity investments, coupled with R&D expenditure, were found to drive above-average revenue generation. This observation was supported by relatively strong model performance, leading to accurate predictions.

## Final Thoughts

While the machine learning models developed in this project demonstrate promising results, they should be considered works in progress requiring further validation with new data and effective deployment. Perhaps the most significant finding is that sustainability investments were the sole metric identified as a KPI across the entire analysis. Simply put, prioritizing an increase in sustainability investments will likely yield the most substantial impact on revenue, ESG score, and market capitalization.
